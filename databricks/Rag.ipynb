```
from databricks.vector_search.client import VectorSearchClient
from langchain_community.vectorstores import DatabricksVectorSearch
from langchain_community.embeddings import DatabricksEmbeddings

embedding_model = DatabricksEmbeddings(endpoint="databricks-bge-large-en")

def get_retriever(persist_dir: str = None):
    os.environ["DATABRICKS_HOST"] = host
    #Get the vector search index
    vsc = VectorSearchClient(workspace_url=host)
    vs_index = vsc.get_index(
        endpoint_name=VECTOR_SEARCH_ENDPOINT_NAME,
        index_name=index_name
    )

    # Create the retriever
    vectorstore = DatabricksVectorSearch(
        vs_index, text_column="text", embedding=embedding_model
    )
    return vectorstore.as_retriever()
```

```
from langchain.chains import RetrievalQA
from langchain_openai import ChatOpenAI
from langchain.prompts import PromptTemplate

# Initialize the OpenAI LLM for GPT-4
llm = ChatOpenAI(model="gpt-4")

# Create a retriever from your vector store
retriever = get_retriever()

# Set up the conversational RAG chain
TEMPLATE = """ 
    {context}
    You are an expert note-taker. Transform the following lecture transcript into well-structured, clear notes.
    Follow these guidelines:
    1. Organize key points into a hierarchical structure
    2. Add bullet points for main ideas
    3. Use proper formatting for headings and subheadings
    4. Correct any grammar or clarity issues
    5. Keep the original meaning intact

    Lecture transcript:
    {transcript}

    Please provide the output as markdown.
Answer:
"""
prompt = PromptTemplate(template=TEMPLATE, input_variables=["context" "transcript"])

# Create the RetrievalQA chain
from langchain_core.runnables import RunnableLambda, RunnablePassthrough
from langchain_core.output_parsers import StrOutputParser
chain = (
    {"context": retriever, "transcript": RunnablePassthrough()}
    | prompt
    | llm
    | StrOutputParser()
)

transcript = '''Today, we are going to discuss data structures, starting with linked lists. A linked list is a sequence of elements, where each element points to the next one...'''
result = chain.invoke(transcript)
print(result)


```

```
from mlflow.models import infer_signature
import mlflow
import langchain

mlflow.set_registry_uri("databricks-uc")
model_name = "quillai.default.note_gen"

with mlflow.start_run(run_name="quillai") as run:
    signature = infer_signature(transcript, result)
    model_info = mlflow.langchain.log_model(
        chain,
        loader_fn=get_retriever,  # Load the retriever with DATABRICKS_TOKEN env as secret (for authentication).
        artifact_path="chain",
        registered_model_name=model_name,
        pip_requirements=[
            "mlflow==" + mlflow.__version__,
            "langchain==" + langchain.__version__,
            "databricks-vectorsearch",
            "langchain_community",
            "langchain-openai"
        ],
        input_example=transcript,
        signature=signature
    )
```
